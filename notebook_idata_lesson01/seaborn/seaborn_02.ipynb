```python
import seaborn as sns
import numpy 
import matplotlib as mpl
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
def sinplot(flip=1):
    x = numpy.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, numpy.sin(x + i + 5) * (7 - i) * flip)

sinplot()
```

```python
sns.set()
sinplot()
```
